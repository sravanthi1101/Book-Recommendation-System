```python
data_df=data_df.drop('Unnamed: 0.1', 1)
```

```python
data_df.rename(columns = {'Unnamed: 0':'Index'}, inplace = True)
data_df.index+=1
data_df.head()
```

```python
booksWithGenres_df = data_df.copy()

#For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
for index, row in data_df.iterrows():
    for genre in row['genre']:
        booksWithGenres_df.at[index, genre] = 1
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
booksWithGenres_df = booksWithGenres_df.fillna(0)
booksWithGenres_df.head()
```

```python
userInput = [
            {'title':'A Thousand Splendid Suns', 'rating':4.7},
            {'title':'The Kite Runner', 'rating':4.5},
            {'title':'The Da Vinci Code', 'rating':3.9}] 
inputdata = pd.DataFrame(userInput)
inputdata
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
```

```python
data_df["combined_text"] = data_df["Desc"] + " " + data_df["author"] + " " + data_df["genre"]
data_df
```

```python
tf = TfidfVectorizer(analyzer = "word", ngram_range=(1,2), min_df=0, stop_words='english')

tfidf_matrix = tf.fit_transform(data_df['combined_text'])

cosine =  cosine_similarity(tfidf_matrix)
print(cosine)
```

```python
Name=data_df['title'][1]
Name
```

```python
book_id=data_df[data_df.title == Name]["Index"].values[0]
book_id
```

```python
similar_books = list(enumerate(cosine[book_id]))
print(similar_books)
```

```python
sortedbooks = sorted(similar_books, key = lambda x:x[1], reverse=True)[1:]
sortedbooks
```
